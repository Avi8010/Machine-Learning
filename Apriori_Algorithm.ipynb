```python
transactions = []
for i in range(0, 7500):
    transactions.append([str(df.values[i,j]) for j in range(0, 20)])
```

```python
rule_list = apriori(transactions, min_support = 0.002, min_confidence = 0.2, min_lift = 3, min_length = 2)
```

```python
res = list(rule_list)
```
